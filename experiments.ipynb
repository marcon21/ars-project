```python
import re
import math
import matplotlib.pyplot as plt
import numpy as np
from tabulate import tabulate
```

***FIRST EXPERIMENT***

First experiment consists in a simulation with low covariance values. R and Q covariances matrixes are reported below. 


1. Diagonal matrix \( R \):
   \[
   R = \begin{pmatrix}
   r_1 & 0 & 0 & \cdots & 0 \\
   0 & r_2 & 0 & \cdots & 0 \\
   0 & 0 & r_3 & \cdots & 0 \\
   \vdots & \vdots & \vdots & \ddots & \vdots \\
   0 & 0 & 0 & \cdots & r_n
   \end{pmatrix}
   \]

2. Diagonal matrix \( Q \):
   \[
   Q = \begin{pmatrix}
   q_1 & 0 & 0 & \cdots & 0 \\
   0 & q_2 & 0 & \cdots & 0 \\
   0 & 0 & q_3 & \cdots & 0 \\
   \vdots & \vdots & \vdots & \ddots & \vdots \\
   0 & 0 & 0 & \cdots & q_n
   \end{pmatrix}
   \]


```python
odd_values = []
even_values = []

with open('./data/low_variance.txt', 'r') as file:
    lines = file.readlines()
    for index, line in enumerate(lines):
        values = re.findall(r'\d+\.*\d*', line)
        values = [float(val) for val in values]
        if index % 2 == 0:
            even_values.append(values[:len(values)//2]) 
            odd_values.append(values[len(values)//2:])  

odd_values_t = list(zip(*odd_values))
even_values_t = list(zip(*even_values))

plt.plot(even_values_t[0], label="x")
plt.plot(odd_values_t[0], label="x_predicted")

plt.title('Graph for values x and x_predicted')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.show()

plt.plot(even_values_t[1], label="y")
plt.plot(odd_values_t[1], label="y_predicted")

plt.title('Graph for values y and y_predicted')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.show()

plt.plot(even_values_t[2], label="theta (degrees)")
plt.plot(odd_values_t[2], label="theta_predicted (degrees)")

plt.title('Graph for values theta and theta_predicted')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.show()
```
