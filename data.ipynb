```python
import re
import matplotlib.pyplot as plt

# Lista per i valori delle righe dispari
odd_values = []

# Lista per i valori delle righe pari
even_values = []

# Lettura del file output.txt
with open('output.txt', 'r') as file:
    lines = file.readlines()
    for index, line in enumerate(lines):
        # Estrae i numeri utilizzando una regex
        values = re.findall(r'\d+\.*\d*', line)
        values = [float(val) for val in values]
        
        # Aggiunge i valori alla lista corretta
        if index % 2 == 0:  # righe pari
            even_values.append(values)
        else:  # righe dispari
            odd_values.append(values)

# Trasposta delle liste per avere v1, v2, v3 separatamente
odd_values_t = list(zip(*odd_values))
even_values_t = list(zip(*even_values))

plt.plot(odd_values_t[0], label="x")
plt.plot(even_values_t[0], label=f"x'")

plt.title('Grafico per righe dispari')
plt.xlabel('Indice')
plt.ylabel('Valore')
plt.legend()
plt.show()

plt.plot(odd_values_t[1], label="y")
plt.plot(even_values_t[1], label=f"y")

plt.title('Grafico per righe pari')
plt.xlabel('Indice')
plt.ylabel('Valore')
plt.legend()
plt.show()


plt.plot(odd_values_t[1], label="theta")
plt.plot(even_values_t[1], label=f"theta''")

plt.title('Grafico per righe pari')
plt.xlabel('Indice')
plt.ylabel('Valore')
plt.legend()
plt.show()
```
