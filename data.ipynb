```python
import re
import matplotlib.pyplot as plt

odd_values = []
even_values = []

with open('output.txt', 'r') as file:
    lines = file.readlines()
    for index, line in enumerate(lines):
        values = re.findall(r'\d+\.*\d*', line)
        values = [float(val) for val in values]
        if index % 2 == 0:
            even_values.append(values[:len(values)//2]) 
            odd_values.append(values[len(values)//2:])  
            
odd_values_t = list(zip(*odd_values))
even_values_t = list(zip(*even_values))
num_plots = len(odd_values_t)
```

```python
plt.plot(odd_values_t[0], label="x")
plt.plot(even_values_t[0], label="x_predicted")

plt.title('Graph for values x and x_predicted')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.show()

plt.plot(odd_values_t[1], label="y")
plt.plot(even_values_t[1], label="y_predicted")

plt.title('Graph for values y and y_predicted')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.show()

plt.plot(odd_values_t[2], label="theta")
plt.plot(even_values_t[2], label="theta_predicted")

plt.title('Graph for values theta and theta_predicted')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.show()
```
